# Solutions of Exercise 6.1

A state-space representation of $G(s) = \frac{1}{s+1}$ is given by

$$
\begin{dcases}
    \dot{x} = \begin{bmatrix} -1 & 0 \\ 0 & 2  \end{bmatrix} x + \begin{bmatrix} 1 \\ 1 \end{bmatrix} u \\
    y = \begin{bmatrix} 1 & 0 \end{bmatrix} x
\end{dcases}
$$ (eq:lec6:ex1_sol:state-space)

1. Compute the poles of {eq}`eq:lec6:ex1_sol:state-space` and compare them with those of $G(s)$. 
2. Is $G(s)$ input-output stable? Is {eq}`eq:lec6:ex1_sol:state-space` asymptotically stable?
3. Examine the controlalbility and observability of the system.
4. Explain why {eq}`eq:lec6:ex1_sol:state-space` is not a good realization of $G(s)$, and provide a better state-space realization.

---

## Solution
### Question 1

The poles of $G(s)$ are the roots of the denominator, i.e., $s = -1$.
The poles of {eq}`eq:lec6:ex1_sol:state-space` are the eigenvalues of $A$, and they can be computed by finding the roots of the characteristic polynomial

$$
\varphi(\lambda) = \det \begin{bmatrix} \lambda + 1 & 0 \\ 0 & \lambda - 2 \end{bmatrix} = (\lambda + 1) (\lambda - 2)
$$

The roots of $\varphi(\lambda) = 0$ are $\lambda = -1$ and $\lambda = 2$.

### Question 2

The system is input-output stable if all the poles of $G(s)$ have negative real part. In this case, we just have one pole in $s=-1$, which means that the system is actually input-output stable.

Asymptotic stability depends, instead, depends on the eigenvalues of $A$ (i.e., on the poles of the state-space system). We have one positive eigenvalue ($\lambda = 2$), which means that the {eq}`eq:lec6:ex1_sol:state-space` is **unstable**.

#### How is this possible?
To understand what's happening, let's assume $x(0) = 0$ and $u(t) = \textrm{step}(t)$. Computing the state trajectory via the Lagrange formula yields

$$
x(t) = \cancel{x(0) e^{At}} + \int_{0}^{t} e^{A\tau} B u(t - \tau) d \tau =  \int_{0}^{t} \begin{bmatrix} e^{-\tau} & 0 \\ 0 & e^{2 \tau} \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix} 1 d \tau 
$$

Since $e^{At}$ is diagonal, solving this integral is faily easy, and can be done independently for the two states:

- $x_1(t) = \int_{0}^{t} e^{-\tau} d \tau = [- e^{-\tau} ]_0^t = 1 - e^{-t}$
- $x_2(t) = \int_{0}^{t} e^{2\tau} d \tau = [0.5 e^{2\tau} ]_0^t = 0.5 e^{2t} - 0.5$

Note that $\lim_{t \to \infty} x_{1}(t) = 1$, meaning that $x_1(t)$ is converging to a constant, while $\lim_{t \to \infty} x_2(t) = + \infty$, meaning that state $x_2(t)$ is diverging. That's why {eq}`eq:lec6:ex1_sol:state-space` is **unstable**.

If we compute, however, $y(t)$, we get

$$
y(t) = 1 x_1(t) + 0 x_2(t) = 1 - e^{-t}
$$

Because $\lim_{t \to \infty} y(t) = 1$ is finite, we can say that the system is input-output stable.

### Question 3

Let's check the controllability matrix:

$$
S = \begin{bmatrix} B & AB \end{bmatrix} = \begin{bmatrix} 1 & -1 \\ 1 & 2 \end{bmatrix}
$$

The system is **controllable**, since $S$ if full rank ($\det(S) = 3 \neq 0$).

Now we move on to the observability matrix:

$$
O = \begin{bmatrix} C \\ CA \end{bmatrix} = \begin{bmatrix} 1 & 0 \\ -1 & 0 \end{bmatrix}
$$

The observabiltiy matrix is singular, as $\det(O) = 0$. Note that the second column is indeed null! This means that the system is **not observable**.

### Question 4

{eq}`eq:lec6:ex1_sol:state-space` is surely not a good realization of $G(s)$ because it has an extra pole ($\lambda = 2$) that is **unstable** and not observable[^obsv].

A much better realization would be any of the **controllable canonical realizations** of $G(s)$, which would give the following first-order state-space model:

$$
\begin{dcases}
\dot{x} = - x + u \\
y = x
\end{dcases}
$$

---

[^obsv]: We can claim that $\lambda = 2$ is the unobservable pole because it is the one not showing up in $G(s)$, and we know that $\det(O) = 0$. To verify this, try to compute $G(s)$ from {eq}`eq:lec6:ex1_sol:state-space` and check that it get cancelled.
