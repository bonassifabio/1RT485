# Exercise 6.0
{ilo}`analysis_model`

Consider the following system:

$$
\begin{dcases}
    \dot{x} = \begin{bmatrix} -2 & 4 \\ 0 & -3  \end{bmatrix} x + \begin{bmatrix} 0 \\ 2 \end{bmatrix} u \\
    y = \begin{bmatrix} 0 & 3 \end{bmatrix} x
\end{dcases}
$$ (eq:lec6:ex0:state-space)

1. Compute the poles of the state-space system {eq}`eq:lec6:ex0:state-space`, and check whether the system is asymptotially stable or not.
2. Is the system controllable and observable?
3. Compute the transfer function $G(s)$ of system {eq}`eq:lec6:ex0:state-space`. 
4. Compute the poles of $G(s)$, and discuss whether they match those of {eq}`eq:lec6:ex0:state-space`.


```{admonition} Inverting $2 \times 2$ matrices
:class: hint
Consider a generic matrix $A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$. Its inverse can be easily computed as

$$
  A^{-1} = \frac{1}{\det(A)} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$ (eq:lec6:ex0:inversion)
```
