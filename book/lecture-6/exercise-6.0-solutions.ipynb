# Solutions of Exercise 6.0

Consider the following system:

$$
\begin{dcases}
    \dot{x} = \begin{bmatrix} -2 & 4 \\ 0 & -3  \end{bmatrix} x + \begin{bmatrix} 0 \\ 2 \end{bmatrix} u \\
    y = \begin{bmatrix} 0 & 3 \end{bmatrix} x
\end{dcases}
$$ (eq:lec6:ex0_sol:state-space)

1. Compute the poles of the state-space system {eq}`eq:lec6:ex0_sol:state-space`, and check whether the system is asymptotially stable or not.
2. Is the system controllable and observable?
3. Compute the transfer function $G(s)$ of system {eq}`eq:lec6:ex0_sol:state-space`. 
4. Compute the poles of $G(s)$, and discuss whether they match those of {eq}`eq:lec6:ex0_sol:state-space`.


```{admonition} Inverting $2 \times 2$ matrices
:class: hint
Consider a generic matrix $A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$. Its inverse can be easily computed as

$$
  A^{-1} = \frac{1}{\det(A)} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$ (eq:lec6:ex0_sol:inversion)
```

---

## Solution
### Question 1

The poles of {eq}`eq:lec6:ex0_sol:state-space` are the eigenvalues of $A$. To compute them, we evaluate the roots of the characteristic polynomial $\varphi(\lambda) = \det (\lambda I - A)$.

$$
\varphi(\lambda) = \det \begin{bmatrix} \lambda + 2 & -4 \\ 0 & \lambda + 3\end{bmatrix} = (\lambda + 2)(\lambda + 3) - (-4) \cdot 0 = (\lambda + 2)(\lambda + 3)
$$

By setting $\varphi(\lambda) = 0$ we get the poles, $\lambda=-2$ and $\lambda=-3$. System {eq}`eq:lec6:ex0_sol:state-space` is **asymptotically stable** because all the poles have negative real part.

### Question 2

To determine whether the system is controllable, we need to check that the controllability matrix is full rank.

$$
S = \begin{bmatrix} B & AB \end{bmatrix} = \begin{bmatrix} 0 & 8 \\ 2 & - 6 \end{bmatrix}
$$

$S$ is full rank. In fact, it's determinant is $\det(S) = 48 \neq 0$.

When it comes to observability, we compute the observability matrix

$$
O = \begin{bmatrix} C \\ CA \end{bmatrix} = \begin{bmatrix} 0 & 3 \\ 0 & -9 \end{bmatrix}
$$

Note that the observability matrix is singular, because the first column is null. Also, $\det(O) = 0$.
Therefore the system is **not (fully) observable**.

### Question 3

To compute $G(s)$, we can apply the formula seen during Lecture 8:

$$
G(s) = C (s I - A)^{-1} B + D
$$

Let's compute it step by step with the help of {eq}`eq:lec6:ex0_sol:inversion`.

$
(sI - A)^{-1} = \begin{bmatrix} s + 2 & -4 \\ 0 & s+3 \end{bmatrix}^{-1} = \frac{1}{(s+2)(s+3)} \begin{bmatrix} s+3 & 4 \\ 0 & s+2 \end{bmatrix}
$

$
C(sI-A)^{-1} = \frac{1}{(s+2)(s+3)} \begin{bmatrix} 0 & 3 \end{bmatrix} \left[\begin{array}{c|c} s+3 & 4 \\ 0 & s+2 \end{array}\right] = \frac{1}{(s+2)(s+3)} \begin{bmatrix} 0 & 3(s+2) \end{bmatrix}
$

$
G(s) = C(sI-A)^{-1} B =  \frac{1}{(s+2)(s+3)} \begin{bmatrix} 0 & 3(s+2) \end{bmatrix} \begin{bmatrix} 0 \\ 2\end{bmatrix} = \frac{6 \cancel{(s+2)}}{\cancel{(s+2)}(s+3)} = \frac{6}{s+3}
$

### Question 4

The only pole of $G(s)$ is $s=-3$. The pole in $s=-2$ has been cancelled while computing $G(s)$, as it was matching a zero of the system (see the last step above).

In this case, we can say that the eigenvalue $\lambda=-2$ is **not observable** (unobservable) because
- It is an eigenvalue of $A$ that doesn't show up in $G(s)$
- The system had one unobservable state ($\textrm{rank}(O) = 2-1 = 1$).

```{admonition} Warning
:class: warning

Pole-zero cancellations can happen when computing the transfer function representation of state-space systems. 
Always remember that

- $\textrm{poles}(G(s)) \subseteq \textrm{eig}(A)$
- Cancellations are especially problematic when the poles being cancelled are unstable!
```
