# Exercise 6.1
{ilo}`analysis_model`

A state-space representation of $G(s) = \frac{1}{s+1}$ is given by

$$
\begin{dcases}
    \dot{x} = \begin{bmatrix} -1 & 0 \\ 0 & 2  \end{bmatrix} x + \begin{bmatrix} 1 \\ 1 \end{bmatrix} u \\
    y = \begin{bmatrix} 1 & 0 \end{bmatrix} x
\end{dcases}
$$ (eq:lec6:ex1:state-space)

1. Compute the poles of {eq}`eq:lec6:ex1:state-space` and compare them with those of $G(s)$. 
2. Is $G(s)$ input-output stable? Is {eq}`eq:lec6:ex1:state-space` asymptotically stable?
3. Examine the controlalbility and observability of the system.
4. Explain why {eq}`eq:lec6:ex1:state-space` is not a good realization of $G(s)$, and provide a better state-space realization.
