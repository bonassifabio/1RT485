# Exercise 6.4
{ilo}`design_time`

Consider the state-space system

$$
\begin{dcases}
\dot{x} = \begin{bmatrix} -1 & -1 \\ 1 & -1 \end{bmatrix} x + \begin{bmatrix} 1 \\ 1 \end{bmatrix} \\
y = \begin{bmatrix} 1 & 0 \end{bmatrix} x
\end{dcases}
$$ (eq:lec6:ex4:state-space)

1. Compute the poles of {eq}`eq:lec6:ex4:state-space`
2. Compute the corresponding transfer function $G(s)$, its poles and zeros.

Design a state-feedback controller $u = -Lx + \ell_0 r$ such that the closed-loop poles are $-2$ and $-3$.

3. Find the state-feedback gain $L$.
4.  Can $\ell_0$ be chosen such that the steady-state error is $0$ when $r$ is constant? Motivate the answer.
